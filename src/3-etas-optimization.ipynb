# 3-etas optimization experiments full input

```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir) 

from qcd.dampingchannels import OneShotEntangledDampingChannel, OneShotEntangledFullInputDampingChannel, OneShotEntangledUniversalDampingChannel
from qcd.configurations import OneShotSetupConfiguration
from qcd import save_object_to_disk
import numpy as np
```

```python
optimization_setup = {'optimizer_algorithms': ['CRS'],
                      'optimizer_iterations': [1000],
                      'eta_partitions': 3,  # number of partitions for the eta ranging from 0 to pi/2
                      'number_channels_to_discriminate': 3,
                      'number_third_channels': 20,
                      'plays': 10000}

filename = '20210421a_C2b_A1_1000_10000_3_6_20'
```

```python
results = OneShotEntangledFullInputDampingChannel.find_optimal_configurations(optimization_setup=optimization_setup)
```

```python
save_object_to_disk(results, name=filename, path="results/")
```

```python
results = OneShotEntangledDampingChannel.find_optimal_configurations(optimization_setup=optimization_setup)
```

```python
save_object_to_disk(results, name=filename, path="results/")
```

```python
two_qubit_full_input_channel = OneShotEntangledFullInputDampingChannel(
    channel_setup_configuration = OneShotSetupConfiguration({
        'points_theta': 20,
        'points_phase': 10,
        'attenuation_angles': np.append(np.arange(0, np.pi/2, np.pi/2/20), np.pi/2)}))
```
