# Testing new structure 

```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir) 

from qcd.dampingchannels import OneShotDampingChannel
from qcd.backends import SimulatorBackend, FakeSimulatorBackend, DeviceIBMQ
from qcd.configurations import OneShotSetupConfiguration
import numpy as np
```

```python
channel = OneShotDampingChannel(channel_setup_configuration = OneShotSetupConfiguration({ 'points_theta': 20,
                                                                                            'points_phase': 10,
                                                                                            'attenuation_factors': np.append(np.arange(0, 1, 0.01), 1)}))
```

```python
channel.plot_first_channel()
```

```python
results = channel.run(backend = SimulatorBackend())
```

```python
results.plot_surface_probabilities()
```

```python
results.plot_wireframe_blochs()
```

```python
channel.plot_fidelity()
```

```python
noisy_channel = OneShotDampingChannel(channel_setup_configuration = OneShotSetupConfiguration({ 'points_theta': 10,
                                                                                                'points_phase': 7,
                                                                                                'attenuation_factors': [0.4]}))
```

```python
noisy_results = noisy_channel.run(backend = [SimulatorBackend(), FakeSimulatorBackend()])
```

```python
noisy_results.plot_wireframe_blochs_one_lambda()
```
